# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado


Referência do pyautogui: https://pyautogui.readthedocs.io/en/latest/quickstart.html

```python
# Instalação 
# pyautogui ->  automações 
# pandas ->  manipulação de dados
# openpyxl ->  leitura de arquivos excel (.xlsx)

# Instalação:  Abrir o terminal e digitar os seguintes comandos: pip install pyautogui pandas openpyxl

# pyautogui.click -> clicar com o mouse
# pyautogui.write -> escrever um texto
# pyautogui.press -> pressionar uma tecla
# pyautogui.hotkey -> pressionar uma combinação de teclas
```

```python
import pyautogui
import time

pyautogui.PAUSE = 1

# Passo a passo do desafio
# Passo 1: Entrar no sistema da empresa (link drive)
# Abrir o Chrome
pyautogui.press("win")
pyautogui.write("chrome")
pyautogui.press("enter")

time.sleep(2)

# Digitei o link do sistema 
link = "https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga"
pyautogui.write(link)

# Apertei enter e esperei 
pyautogui.press("enter")
time.sleep(3)

# Passo 2: Navegar no sistema para encontrar a base de dados 
pyautogui.click(x=410, y=365, clicks=2)
time.sleep(1)


# Passo 3: Exportar a base de dados (baixar o arquivo)
pyautogui.click(x=461, y=481) # Seleciona o arquivo
pyautogui.click(x=505, y=368) # Clica nos 3 pontinhos
pyautogui.click(x=537, y=444) # Clica em download
time.sleep(3)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
import pandas


# Passo 4: Calcular os indicadores (faturamento, quantidade de produtos vendidos)
# Abrir a base de dados com o Excel
caminho= r"C:\Users\Maria Luiza\Downloads\Vendas - Dez.xlsx"
tabela = pandas.read_excel(caminho) 
# Ver as informações da base de dados
display(tabela)
# Somar o faturamento de todos os produtos
faturamento = tabela["Valor Final"].sum()
# Somar a quantidade de produtos
qtde_produtos = tabela["Quantidade"].sum()
print(qtde_produtos)
```

### Vamos agora enviar um e-mail pelo gmail

```python
import pyperclip

# Passo 5: Enviar as informações por e-mail
# Abrir um nova aba no Chrome
pyautogui.hotkey("ctrl", "t")  # Abrir uma nova aba


time.sleep(2)
pyautogui.write("https://mail.google.com/")
pyautogui.press("enter")
time.sleep(5)

pyautogui.click(x=107, y=223)  # Clicar em escrever
time.sleep(5)

# Para quem enviar o email
pyautogui.write("ml9634100@gmail.com")
pyautogui.press("tab")  # Seleciona o email
pyautogui.press("tab")  # Vai para o campo de assunto

# Qual o assunto do email
pyperclip.copy("Relatório de vendas")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab")  # Vai para o campo do corpo do email

# Qual o corpo do email
texto = f"""
Prezado, 
Segue o relatório de vendas de hoje:
Faturamento: R${faturamento}
Quantidade de produtos vendidos: {qtde_produtos}

Qualquer dúvida, estou à disposição.
Deseja mais alguma coisa? 

Abs,
Maria Luiza
"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

# Enviar o email
pyautogui.hotkey("ctrl", "enter")  # Enviar usando atalho do Gmail
```
